### Low frequency bins (band A) show large variations.

```python
from scipy.io.idl import readsav
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
m_dat=readsav('/Users/tongyuguang/Google Drive/research/data/meudon_tnr/TNR_XY_ACE_19950610.sav')
```

```python
m_data=m_dat['data']
mt = m_data['timeur8'][0]
mtag = m_data['time'][0]
spec = m_data['spectra'][0]
spec = 10. ** (spec/10.)
```

```python
fbins = np.array([4000*1.0445**i for i in range(96)])
```

```python
fig = plt.figure(figsize=[6, 6])
plt.plot(fbins, spec[:, 110:120])
plt.xscale('log')
plt.yscale('log')
plt.xlim([4e3, 256e3])
plt.ylim([1e-16, 5e-13])
plt.xlabel(r'$f[kHz]$')
plt.ylabel(r'$V_r^2[V^2Hz^{-1}]$')
plt.title('10 spectra near ' + mtag[115].decode("utf-8"))
plt.show()
```

```python
fig.savefig('wind1.png', dpi=300)
```
