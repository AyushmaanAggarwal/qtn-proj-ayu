## Finding the best huber value

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Format: list of lists that contain 
#[huber_delta, median, mean, percent75] for percent error
data = [[.5, 1.266867e-02, 1.906012e-02, 2.342065e-02],
        [.75, ]
        [1,  8.188022e-03, 1.449772e-02, 1.619062e-02],
        [2,  9.602736e-03, 1.525508e-02, 1.903147e-02],
        [5,  1.183179e-02, 1.737781e-02, 2.220664e-02],
        [10, 9.480028e-03, 1.556152e-02, 1.886097e-02],
        [15, 9.480028e-03, 1.556152e-02, 1.886097e-02],
        [20, 9.480028e-03, 1.556152e-02, 1.886097e-02]]

huber_delta = list(zip(*data))[0]
median      = list(zip(*data))[1]
mean        = list(zip(*data))[2]
perc75      = list(zip(*data))[3]
```

```python
plt.plot(huber_delta, median, label="Median Percent Error")
plt.plot(huber_delta, mean, label="Mean Percent Error")
plt.plot(huber_delta, perc75, label="75th Percentile for Percent Error")
plt.title("Performance over different delta huber loss functions")
plt.xlabel("Huber Delta Value")
plt.ylabel("Metrics of Percent Error")
plt.xlim(0, 21)
plt.ylim(0, .025)
plt.legend()
```
